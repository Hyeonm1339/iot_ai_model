```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_file = '/home/workspace/iot_ai_model/dataset/digital_digit/anno_digit.txt'
```

```python
with open(a_file, 'r') as f:
    annotation = f.readlines()
```

```python
cols = annotation[0].strip().split()
filename = cols[0]
objects = cols[1:]
```

```python
filename
```

```python
obj = objects[0]
box = np.array(obj.split(','), dtype = np.int)
```

```python
img = cv2.imread('dataset/digital_digit/' + filename)
```

```python
output_img = cv2.rectangle(img, (box[0], box[1]), (box[2], box[3]), (0,0,255), 2)
```

```python
output_img = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.imshow(output_img)
plt.show()
```
